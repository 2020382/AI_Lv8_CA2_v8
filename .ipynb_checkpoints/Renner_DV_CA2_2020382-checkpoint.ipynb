### Differences between a graph and a tree !


Graph is a collection of two sets also known as nodes and edges that connect the two sets, graph can be direct and undirect, it just means that edges can have a specific direction or not.

A directed graph can also have weighted edges, and is also known as a digraph, where each edge has a weight or cost associated with it. Graphs can be represented in various ways such as adjacency matrix or adjacency list. In a direct graph the edges/nodes are linked to one direction whereas undirected can have diferent paths not only one forward way like direct graph.


A tree is a non-linear data structure that represents the hierarchy. A tree is a collection of nodes that are linked together to form a hierarchy and its elements are arranged in multiple levels. The figure below is an example of a hierachy tree.

### Implementation of Dijkstra's algorithm

```python
import heapq

def dijkstra(graph, start):
    
    # We initialise the distances with infinity for all nodes and 0 to the start node which is A.
    distances = {node: float('infinity') for node in graph}
    distances[start] = 0

    # The priority queue keep track of what is the next node to visit
    priority_queue = [(0, start)]

    # From the priority queue we choose the node with the shortest distance
    while priority_queue:
        current_distance, current_node = heapq.heappop(priority_queue)

        # Verifies if the path to the current node is longer than the current path 
        if current_distance > distances[current_node]:
            continue

        # Check neighbors
        for neighbor, weight in graph[current_node].items():
            distance = current_distance + weight

            # If we find a shorter path, we update the distance and push it to the priority queue
            if distance < distances[neighbor]:
                distances[neighbor] = distance
                heapq.heappush(priority_queue, (distance, neighbor))

    return distances

# Here goes the puzzle graph.
puzzle_graph = {
   'A': {'B': 1, 'C': 5, 'H': 2},
    'B': {'D': 4, 'C': 2, 'A': 1},
    'C': {'B': 2, 'G': 1},
    'D': {'B': 4, 'F': 7, 'L':7},
    'E': {'D': 4,'F': 3,'W': 6},
    'F': {'G': 2, 'D': 7, 'E': 3},
    'G': {'C': 1, 'F': 2, 'L': 3},
    'H': {'A': 2, 'K': 5, 'J': 9},
    'J': {'H': 9, 'N': 6, 'K': 3},
    'K': {'L': 5, 'H': 5, 'J': 3},
    'L': {'K': 5, 'N': 3, 'M': 4, 'V': 10, 'W': 8, 'G': 3, 'D': 7},
    'M': {'L': 4, 'P': 2, 'Q': 10},
    'N': {'J': 6, 'P': 4, 'L': 3, 'S': 7},
    'P': {'N': 4, 'R': 5, 'M': 2},
    'Q': {'M': 10, 'W': 4, 'S': 8},
    'R': {'S': 4, 'T': 3, 'P': 5},
    'S': {'R': 4, 'T': 4, 'U': 2, 'V':6, 'Q': 8, 'N': 7},
    'T': {'R': 3, 'S': 4, 'U': 1},
    'U': {'S': 2, 'T': 1, 'V': 3},
    'V': {'W': 5, 'S': 6, 'U': 3, 'L': 10},
    'W': {'E': 6, 'V': 5, 'Q': 4, 'L': 8},

}

# Calculate and print the distance using dijkstra algorithim
distances_dijkstra = dijkstra(puzzle_graph, 'A')
print(distances_dijkstra)
```

### Implementation of A* algorithm

```python
import heapq

def astar(graph, start, goal):
    
    # Heuristic function
    def heuristic(node):
        
        # the heuristic is zero for now
        return 0

    # Initialize with infinity costs and heuristics for all nodes except the starting point A set to 0
    costs = {node: float('infinity') for node in graph}
    costs[start] = 0

    # Keep track of the next node to visit
    priority_queue = [(0 + heuristic(start), 0, start)]

    while priority_queue:
        _, current_cost, current_node = heapq.heappop(priority_queue)

        # Check if the path to the current node is already longer than the the known path
        if current_cost > costs[current_node]:
            continue

        # once the goal is achieved it will stop the loop
        if current_node == goal:
            break

        # Explore neighbors
        for neighbor, weight in graph[current_node].items():
            cost = current_cost + weight
            heuristic_value = heuristic(neighbor)

            # If a shorter path is found, update the cost to the priority queue
            if cost + heuristic_value < costs[neighbor]:
                costs[neighbor] = cost
                heapq.heappush(priority_queue, (cost + heuristic_value, cost, neighbor))

    return costs

# graph
puzzle_graph = {
    
    'A': {'B': 1, 'C': 5, 'H': 2},
    'B': {'D': 4, 'C': 2, 'A': 1},
    'C': {'B': 2, 'G': 1},
    'D': {'B': 4, 'F': 7, 'L':7},
    'E': {'D': 4,'F': 3,'W': 6},
    'F': {'G': 2, 'D': 7, 'E': 3},
    'G': {'C': 1, 'F': 2, 'L': 3},
    'H': {'A': 2, 'K': 5, 'J': 9},
    'J': {'H': 9, 'N': 6, 'K': 3},
    'K': {'L': 5, 'H': 5, 'J': 3},
    'L': {'K': 5, 'N': 3, 'M': 4, 'V': 10, 'W': 8, 'G': 3, 'D': 7},
    'M': {'L': 4, 'P': 2, 'Q': 10},
    'N': {'J': 6, 'P': 4, 'L': 3, 'S': 7},
    'P': {'N': 4, 'R': 5, 'M': 2},
    'Q': {'M': 10, 'W': 4, 'S': 8},
    'R': {'S': 4, 'T': 3, 'P': 5},
    'S': {'R': 4, 'T': 4, 'U': 2, 'V':6, 'Q': 8, 'N': 7},
    'T': {'R': 3, 'S': 4, 'U': 1},
    'U': {'S': 2, 'T': 1, 'V': 3},
    'V': {'W': 5, 'S': 6, 'U': 3, 'L': 10},
    'W': {'E': 6, 'V': 5, 'Q': 4, 'L': 8},
}

# Start and goal nodes
start_node = 'A'
goal_node = 'S'

# Calculate costs using A* algorithm
costs_astar = astar(puzzle_graph, start_node, goal_node)
print(costs_astar)
```

### Timing Analysis

#### Comparing the actual time it takes for each algorithm to find the solution, using time module with djikstra and A*

```python
import time
import heapq
import math

def dijkstra_algorithm(graph, start, goal):
    distances = {node: float('infinity') for node in graph}
    distances[start] = 0
    priority_queue = [(0, start)]

    while priority_queue:
        current_distance, current_node = heapq.heappop(priority_queue)

        if current_node == goal:
            break

        for neighbor, weight in graph[current_node].items():
            distance = current_distance + weight
            if distance < distances[neighbor]:
                distances[neighbor] = distance
                heapq.heappush(priority_queue, (distance, neighbor))

    return distances[goal]

def astar_algorithm(graph, start, goal, heuristic_function):
    priority_queue = [(0, start)]
    g_values = {node: float('infinity') for node in graph}
    g_values[start] = 0

    while priority_queue:
        current_g, current_node = heapq.heappop(priority_queue)

        if current_node == goal:
            break

        for neighbor, weight in graph[current_node].items():
            tentative_g = g_values[current_node] + weight
            if tentative_g < g_values[neighbor]:
                g_values[neighbor] = tentative_g
                f_value = tentative_g + heuristic_function(neighbor, goal)
                heapq.heappush(priority_queue, (f_value, neighbor))

    return g_values[goal]

# Graph information
puzzle_graph = {
    
    'A': {'B': 1, 'C': 5, 'H': 2},
    'B': {'D': 4, 'C': 2, 'A': 1},
    'C': {'B': 2, 'G': 1},
    'D': {'B': 4, 'F': 7, 'L':7},
    'E': {'D': 4,'F': 3,'W': 6},
    'F': {'G': 2, 'D': 7, 'E': 3},
    'G': {'C': 1, 'F': 2, 'L': 3},
    'H': {'A': 2, 'K': 5, 'J': 9},
    'J': {'H': 9, 'N': 6, 'K': 3},
    'K': {'L': 5, 'H': 5, 'J': 3},
    'L': {'K': 5, 'N': 3, 'M': 4, 'V': 10, 'W': 8, 'G': 3, 'D': 7},
    'M': {'L': 4, 'P': 2, 'Q': 10},
    'N': {'J': 6, 'P': 4, 'L': 3, 'S': 7},
    'P': {'N': 4, 'R': 5, 'M': 2},
    'Q': {'M': 10, 'W': 4, 'S': 8},
    'R': {'S': 4, 'T': 3, 'P': 5},
    'S': {'R': 4, 'T': 4, 'U': 2, 'V':6, 'Q': 8, 'N': 7},
    'T': {'R': 3, 'S': 4, 'U': 1},
    'U': {'S': 2, 'T': 1, 'V': 3},
    'V': {'W': 5, 'S': 6, 'U': 3, 'L': 10},
    'W': {'E': 6, 'V': 5, 'Q': 4, 'L': 8},
}

# Start and goal nodes, set A as starting point and S the destination we want to reach
start_node = 'A'
goal_node = 'S'

# Heuristic funciont to calculate our shortest path
def heuristic_function(node, goal):
    # Euclidean distance as a heuristic
    x1, y1 = node_coordinates[node]
    x2, y2 = node_coordinates[goal]
    return math.sqrt((x2 - x1)**2 + (y2 - y1)**2)

# Coordinates of each node and its value
node_coordinates = {
    'A': (0, 0),
    'B': (1, 1),
    'C': (2, 0),
    'D': (3, 2),
    'E': (4, 1),
    'F': (5, 2),
    'G': (6, 0),
    'H': (1, 3),
    'J': (2, 4),
    'K': (3, 3),
    'L': (4, 4),
    'M': (5, 3),
    'N': (6, 4),
    'P': (7, 3),
    'Q': (8, 4),
    'R': (9, 3),
    'S': (5, 5),
    'T': (9, 4),
    'U': (8, 2),
    'V': (7, 5),
    'W': (6, 6),
}

# Measure time for Dijkstra's algorithm
start_time_dijkstra = time.time()
dijkstra_solution = dijkstra_algorithm(puzzle_graph, start_node, goal_node)
end_time_dijkstra = time.time()
time_taken_dijkstra = end_time_dijkstra - start_time_dijkstra

# Measure time for A* algorithm
start_time_astar = time.time()
astar_solution = astar_algorithm(puzzle_graph, start_node, goal_node, heuristic_function)
end_time_astar = time.time()
time_taken_astar = end_time_astar - start_time_astar

# Print the results
print(f"Shortest distance from {start_node} to {goal_node} using Dijkstra's: {dijkstra_solution}")
print(f"Time taken for Dijkstra's algorithm: {time_taken_dijkstra}")

print(f"Shortest distance from {start_node} to {goal_node} using A* Algorithm: {astar_solution}")
print(f"Time taken for A* algorithm: {time_taken_astar}")
```
